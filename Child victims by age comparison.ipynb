```python
#Since normalilty is not satisfied, options to analysis data would either be nonparametric ANOVA method (Kruskal Wallis test) 
#or transformation method (logx)

#using Kruskal Wallis test
#H0:mu1 = mu2 = .. = mu17 (mean ranks of all age group are equal)
#HA:at least one age group mean rank is not equal 

#TS:
stats.kruskal(age0,age1,age2,age3,age4,age5,age6,age7,age8,age9,age10,age11,age12,age13,age14,age15,age16,age17,age18)
#df = len(ages) - 1
#df
#RR: (alpha = 0.05)

#Conclusion:
#Based on the data, there is sufficient evidence that the mean ranks for the age groups are not all equal and there would be at least one
#age group with a different mean on average.
```

```python
#post hoc test (dunns)

#put ages into a list
ages = [age0, age1, age2, age3, age4, age5, age6, age7, age8, age9, age10, age11, age12, age13, age14, age15, age16, age17, age18]

sp.posthoc_dunn(ages, p_adjust = 'bonferroni')
```

```python
#Post-Hoc Analysis
#Based on Dunn's test with Bonferroni adjustment, a lot of the ages have a very simliar number of victims. 
#However, there there seems to be a trend in which children that are younger, when comparing to older victims have more of a difference than with victims closer to
#their age (ex: age 0 vs age 5 would be closer in number of victims than age0 vs age 17)
#There is one age group (age 18-21) that is very different from all the other ones. Even when comparing to the next closest age group (age 17), the p-value is still tiny.
#There could be outside factors such as being 18 - 21, those victims would be seen as adults and might not be targeted as much as the younger age groups or families
#don't report these issues as the victim might be independent from thier families.
#In conclusion, based on the data, children that are younger have a higher chance of being victims.
```
